```python
import requests
from requests.structures import CaseInsensitiveDict

url = 'https://floral-sparkling-lake.discover.quiknode.pro/3345460a18b30a5f499577660fd08945d6377d55/'

headers = CaseInsensitiveDict()
headers["Content-Type"] = "application/json"
headers["x-qn-api-version"] = "1"

# first request
data = """
{
    "id":67,
    "jsonrpc":"2.0",
    "method":"qn_fetchNFTsByCollection",
    "params":{
      "collection": "0xa7d8d9ef8D8Ce8992Df33D8b8CF4Aebabd5bD270",
    "omitFields": ["description"],
      "page": 1,
      "perPage": 100
    }
  }
"""
resp = requests.post(url, headers=headers, data=data)
print(resp.status_code)
dt = resp.json()

# Get number of pages
n = dt['result']['totalPages']

# Get results into list
data_list = [dt['result']]
```

```python
def get_transactions(n,headers,url,data_list):
    a = []
    for i in range(2,n+1):
        
        data = """
        {
            "id":67,
            "jsonrpc":"2.0",
            "method":"qn_fetchNFTsByCollection",
            "params":{
              "collection": "0xa7d8d9ef8D8Ce8992Df33D8b8CF4Aebabd5bD270",
            "omitFields": ["description"],
              "perPage": 100,
              "page": """ + str(i) +"""}}"""
        
        #print(data)
        
        resp = requests.post(url, headers=headers, data=data)
        print(i,resp.status_code)
        dt = resp.json()
        
        #data_list = 
        a.append(dt['result'])
        
        
    return a        
```

```python
raw_data_ = get_transactions(n,headers,url,data_list)

raw_data = data_list + raw_data_
```

```python
# Get data into dataframe
full = []
for i in range(0,len(raw_data)):
    #print(i)
    df_ = pd.DataFrame(raw_data[i]['tokens'])
    full.append(df_)
df = pd.concat(full)
df.reset_index(drop=True,inplace=True)

df.fillna('unknown',inplace=True)
#df.replace('[ ]','unknown',inplace=True)
df
```

```python
# Get price for transactions
def get_eth(url,tx):
    headers = CaseInsensitiveDict()
    headers["Content-Type"] = "application/json"
    
    data = '''{"method":"eth_getTransactionByHash","id":1,"jsonrpc":"2.0",
    "params": ["''' + str(tx) + '''"]}'''

    resp = requests.post(url, headers=headers, data=data)

    #print(resp.status_code)
    
    dt = resp.json()

    # Get results into list
    data_list = [dt]
    
    return data_list
```

```python
# Get prices by transaction
txn = []
for i in range(df_provenance.index.min(),df_provenance.index.max()+1):
    print(i)
    txn_ = get_eth(url,df_provenance.txHash[i])
    txn_ = txn_[0]['result']
    txn__ = pd.DataFrame([txn_])

    # Decode value and gas price
    txn__.value = txn__.value.apply(lambda x: int(x, 16))
    txn__.value = txn__.value/1000000000000000000

    txn__.gasPrice = txn__.gasPrice.apply(lambda x: int(x, 16))
    txn__.gasPrice = txn__.gasPrice/1000000000000000000    
    
    txa = txn__[['value','gasPrice']]
    txa['txHash'] = df_provenance.txHash[i]
    
    txn.append(txa)

df_txn = pd.concat(txn)
df_txn.reset_index(drop=True,inplace=True)   
df_txn
```
