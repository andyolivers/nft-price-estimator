```python
import requests
import pandas as pd
import numpy as np
```

```python


url = "https://api.nftgo.dev/eth/v1/market/chart/marketcap?start_time=1577923200&end_time=1659139200"

headers = {
    "Accept": "application/json",
    "X-API-KEY": "edc2cd91f4e34782921c46ec60e7db74"
}

response = requests.get(url, headers=headers)

print(response.text)
```

```python
dt = response.json()
df = pd.DataFrame(dt)
df['x'] = pd.to_datetime(df['x'],unit='s')
df.sort_values(by='x', ascending=True,inplace=True)

from datetime import date, timedelta


date = pd.date_range(df['x'][0],df['x'][93]-timedelta(days=1),freq='d')
df_date = pd.DataFrame(data=date, columns=['DATE'])

df_ = df_date.merge(df, how='left', left_on='DATE', right_on='x', indicator=False)

df_.fillna(method="ffill",inplace=True)
```

```python
df_['market_cap'] =df_['y']

df_[['DATE','market_cap']].to_excel('market_cap.xlsx')
```
