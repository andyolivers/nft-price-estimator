```python
# Load data
#path = '/Users/andreoliveira/Desktop/NOVA IMS/Courses/Tese/data/market/nonfungible/'
path = '/Users/andreoliveira/Desktop/NOVA IMS/Courses/Tese/data/market/google trends/'

chart_0 = pd.read_csv(path+'art-blocks-multiTimeline.csv', delimiter= ',',index_col=None, header=1)
chart_1 = pd.read_csv(path+'blockchain-multiTimeline.csv', delimiter= ',',index_col=None, header=1)
chart_2 = pd.read_csv(path+'eth-multiTimeline.csv', delimiter= ',',index_col=None, header=1)
chart_3 = pd.read_csv(path+'metaverse-multiTimeline.csv', delimiter= ',',index_col=None, header=1)
chart_4 = pd.read_csv(path+'non-fungible-token-multiTimeline.csv', delimiter= ',',index_col=None, header=1)
```

```python
# Join dataframes
chart_01 = chart_0.merge(chart_1, how='inner', on='Semana', indicator=False)
chart_012 = chart_01.merge(chart_2, how='inner', on='Semana', indicator=False)
chart_0123 = chart_012.merge(chart_3, how='inner', on='Semana', indicator=False)
chart_01234 = chart_0123.merge(chart_4, how='inner', on='Semana', indicator=False)

chart_01234
```

```python
chart_01234.replace('<1', 1, inplace=True)
chart_01234['INTEREST_SCORE'] = chart_01234[['art blocks: (A nível mundial)',
       'blockchain: (A nível mundial)', 'Ethereum: (A nível mundial)',
       'Metaverso: (A nível mundial)',
       'Token não fungível: (A nível mundial)']].mean(axis=1)
```

```python
chart_01234
```

```python
from datetime import date, timedelta

chart_01234['Semana'] = pd.to_datetime(chart_01234['Semana'])


date = pd.date_range(chart_01234['Semana'][0],chart_01234['Semana'][259]-timedelta(days=1),freq='d')
df_date = pd.DataFrame(data=date, columns=['DATE'])
```

```python
chart_01234_ = df_date.merge(chart_01234, how='left', left_on='DATE', right_on='Semana', indicator=False)
```

```python
chart_01234_.fillna(method="ffill",inplace=True)
```

```python
chart_01234_final = chart_01234_[['DATE','INTEREST_SCORE']]

chart_01234_final.to_excel('google_trends.xlsx')
```
